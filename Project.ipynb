```python
target = 'temp'
features = df.drop(columns=['temp', 'conditions', 'conditions_encoded'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, df[target], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.cluster import KMeans
X_cluster = df[['temp', 'humidity', 'heat_index']]
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster_kmeans'] = kmeans.fit_predict(X_cluster)
```

```python
sns.scatterplot(data=df, x='temp', y='humidity', hue='cluster_kmeans', palette='viridis')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(df.select_dtypes(include=['float64','int64']))
pca = PCA(n_components=2)
pca_vals = pca.fit_transform(scaled)
df['pca1'] = pca_vals[:,0]
df['pca2'] = pca_vals[:,1]
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20]
}
grid = GridSearchCV(RandomForestClassifier(), param_grid, cv=3, scoring='accuracy')
grid.fit(X_train, y_train)

print("Best params:", grid.best_params_)
print("Best accuracy:", grid.best_score_)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# KMeans
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(X_cluster)
sil_score = silhouette_score(X_cluster, clusters)

# PCA
scaler = StandardScaler()
scaled = scaler.fit_transform(X_cluster)

pca = PCA(n_components=2)
pca_values = pca.fit_transform(scaled)

explained_var = pca.explained_variance_ratio_

print("KMeans Silhouette Score:", sil_score)
print("PCA Explained Variance:", explained_var)
```
